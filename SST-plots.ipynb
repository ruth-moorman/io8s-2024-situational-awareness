This notebook uses the [Copernicus Marine Toolbox](https://help.marine.copernicus.eu/en/articles/7970514-copernicus-marine-toolbox-installation) to download a subset of Copernicus's [ODYSSEA Global Sea Surface Temperature Gridded Level 4 Daily Multi-Sensor Product](https://data.marine.copernicus.eu/product/SST_GLO_PHY_L4_NRT_010_043/description) for the region of our IO8S 2024 cruise for an input date and plots said SST data.

**From the test run pre cruise, it looks like data from (day) GMT is available on (day+2) GMT, between a 1 and 2 day lag for processing.**

```python
import copernicusmarine #note - need to set up an account
import numpy as np
from matplotlib import pyplot as plt 
import cmocean.cm as cm
import xarray as xr
```

```python
# get data for the full cruise domain
dataset_id = "cmems_obs-sst_glo_phy_nrt_l4_P1D-m"
# variables=["sla", "flag_ice", "ugos", "vgos"]
minimum_longitude=70
maximum_longitude=120
minimum_latitude=-70
maximum_latitude=-25
```

## downloading a day of data (updating your data directory)

```python
## modify for target day
today_string = "2024-02-10"
```

```python
start_datetime = today_string+"T00:00:00"
end_datetime   = today_string+"T23:59:59"
filename       = "copernicus_SST_IO8S_multisat"+today_string+".nc"
```

```python
copernicusmarine.subset(
  dataset_id=dataset_id,
#   variables=variables,
  minimum_longitude = minimum_longitude,
  maximum_longitude = maximum_longitude,
  minimum_latitude  = minimum_latitude,
  maximum_latitude  = maximum_latitude,
  start_datetime    = start_datetime,
  end_datetime      = end_datetime,
  minimum_depth=0,
  maximum_depth=1,
  output_filename = filename,
  output_directory = "copernicus-data",
  username = 'rmoorman1',
  password = 'IO8S-cruise'
)
```

```python
def box(ax,W,E,S,N,color):
    ax.plot([W,W],[S,N], color, linewidth=2)
    ax.plot([E,E],[S,N], color, linewidth=2)
    ax.plot([W,E],[N,N], color, linewidth=2)
    ax.plot([W,E],[S,S], color, linewidth=2)
```

```python
fig, (ax, ax1) = plt.subplots(1,2, figsize=(21,10), gridspec_kw={'width_ratios': [1, 1], 'wspace':0.28})

w=3
c = ax.pcolormesh(ds.longitude, ds.latitude,ds.analysed_sst.isel(time=0)-273.15, cmap=cm.thermal, vmin=-2, vmax=25)
ax.contour(ds.longitude, ds.latitude,ds.sea_ice_fraction.isel(time=0), levels=[0,0.15], colors=['limegreen']) # flag ice 15% ice criterion
ax.contour(ds.longitude, ds.latitude,ds.sea_ice_fraction.isel(time=0), levels=[0.05], colors=['limegreen']) # flag ice 15% ice criterion
ax.plot(lon, lat, 'v', color='yellow', markersize=10, markeredgecolor='k')
box(ax,lon-w,lon+w,lat-w,lat+w,'k')
ax.set_title(str(ds.analysed_sst.isel(time=0).time.values)[:10])

cax  = fig.add_axes([0.49, 0.11, 0.02, 0.77])
cbar = plt.colorbar(c,cax = cax,orientation='vertical')
cbar.set_label('SST [C]', fontsize = 15)


x = ds.longitude.sel(longitude = slice(lon-w,lon+w))
y = ds.latitude.sel(latitude = slice(lat-w,lat+w))
sst = ds.analysed_sst.isel(time=0).sel(longitude = slice(lon-w,lon+w), latitude = slice(lat-w,lat+w))-273.15
c = ax1.contourf(x, y,sst, cmap=cm.thermal, levels = np.arange(np.ceil(sst.min().values),np.floor(sst.max().values),0.25), extend='both')
# c = ax1.pcolormesh(x, y,sst, cmap=cm.thermal, vmin=sst.min().values, vmax=sst.max().values)
ax1.plot(lon, lat, 'v', color='yellow', markersize=20, markeredgecolor='k')

cax  = fig.add_axes([0.92, 0.11, 0.02, 0.77])
cbar = plt.colorbar(c,cax = cax,orientation='vertical')
cbar.set_label('SST [C]', fontsize = 15)

plt.show()
```

## plottind today's data and the preceding 3 days of data

```python
filename1 = "copernicus-data/copernicus_SST_IO8S_multisat"+today_string+".nc"
filename2 = "copernicus-data/copernicus_SST_IO8S_multisat"+str(np.datetime64(today_string)-1)+".nc"
filename3 = "copernicus-data/copernicus_SST_IO8S_multisat"+str(np.datetime64(today_string)-2)+".nc"
filename4 = "copernicus-data/copernicus_SST_IO8S_multisat"+str(np.datetime64(today_string)-3)+".nc"
```

```python
ds = xr.open_mfdataset({filename1, filename2, filename3, filename4})
```

```python
w=3;
## reference location (e.g. ship location), for zoomed plotstime
lon=lon;
lat=lat;


levels = np.arange(np.ceil((ds.analysed_sst.sel(longitude = slice(lon-w,lon+w), latitude = slice(lat-w,lat+w))-273.15).min().values),np.floor((ds.analysed_sst.sel(longitude = slice(lon-w,lon+w), latitude = slice(lat-w,lat+w))-273.15).max().values),0.25)

for i in range(4):
    fig, (ax, ax1) = plt.subplots(1,2, figsize=(21,10), gridspec_kw={'width_ratios': [1, 1], 'wspace':0.28})
    ## full domain
    c = ax.pcolormesh(ds.longitude, ds.latitude,ds.analysed_sst.isel(time=i)-273.15, cmap=cm.thermal, vmin=-2, vmax=25)
    ax.contour(ds.longitude, ds.latitude,ds.sea_ice_fraction.isel(time=i), levels=[0,0.15], colors=['limegreen']) # flag ice 15% ice criterion
    ax.plot(lon, lat, 'v', color='yellow', markersize=10, markeredgecolor='k')
    box(ax,lon-w,lon+w,lat-w,lat+w,'k')
    ax.set_title(str(ds.analysed_sst.isel(time=i).time.values)[:10])

    cax  = fig.add_axes([0.49, 0.11, 0.02, 0.77])
    cbar = plt.colorbar(c,cax = cax,orientation='vertical')
    cbar.set_label('SST [C]', fontsize = 15)


    x = ds.longitude.sel(longitude = slice(lon-w,lon+w))
    y = ds.latitude.sel(latitude = slice(lat-w,lat+w))
    sst = ds.analysed_sst.isel(time=i).sel(longitude = slice(lon-w,lon+w), latitude = slice(lat-w,lat+w))-273.15
    c = ax1.contourf(x, y,sst, cmap=cm.thermal, levels = levels, extend='both')
    # c = ax1.pcolormesh(x, y,sst, cmap=cm.thermal, vmin=sst.min().values, vmax=sst.max().values)
    ax1.plot(lon, lat, 'v', color='yellow', markersize=20, markeredgecolor='k')

    cax  = fig.add_axes([0.92, 0.11, 0.02, 0.77])
    cbar = plt.colorbar(c,cax = cax,orientation='vertical')
    cbar.set_label('SST [C]', fontsize = 15)

plt.show()
```
